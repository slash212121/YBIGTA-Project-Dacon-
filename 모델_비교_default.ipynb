```
old_train=pd.read_csv('/content/gdrive/My Drive/train.csv')
old_test=pd.read_csv('/content/gdrive/My Drive/test.csv')
new_train=pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/ybigta_sdss_train.csv')
new_test=pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/ybigta_sdss_test.csv')
print(old_train.head())
print(new_train.head())
```

```
from catboost import CatBoostClassifier
```

```
cb=CatBoostClassifier()
```

```
old_X=old_train.drop(columns=['type'],axis=1)
old_y=old_train['type']
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(old_X,old_y,test_size=0.2,random_state=21) 
```

```
cb.fit(X_train,y_train)
```

```
predictions=cb.predict_proba(X_test)
predictions
```

```
from sklearn.metrics import log_loss
log_loss(y_true=y_test,y_pred=predictions)
```

```
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
```

```
rf.fit(X_train,y_train)
```

```
predictions_rf=rf.predict_proba(X_test)
log_loss(y_true=y_test,y_pred=predictions_rf)
```

```
from xgboost import XGBClassifier
xgb=XGBClassifier()
```

```
xgb.fit(X_train,y_train)
```

```
predictions_xgb=xgb.predict_proba(X_test)
log_loss(y_true=y_test,y_pred=predictions_xgb)
```

```
from lightgbm import LGBMClassifier
lgbm=LGBMClassifier()
```

```
lgbm.fit(X_train,y_train)
```

```
predictions_lgbm=lgbm.predict_proba(X_test)
log_loss(y_true=y_test,y_pred=predictions_lgbm)
```

```
train_for_scaling=old_train.drop(columns=['id','type','fiberID'],axis=1)
from sklearn.preprocessing import RobustScaler
rbscaler=RobustScaler()
print(rbscaler.fit(train_for_scaling))
train_robustscaled = rbscaler.transform(train_for_scaling) #robustscaling 적용
train_robustscaled = pd.DataFrame(train_robustscaled,columns=train_for_scaling.columns)
train_robustscaled = pd.concat([old_train[['id','type', 'fiberID']], train_robustscaled], axis=1) 
train_robustscaled
```

```
test_for_scaling=old_test.drop(columns=['id','fiberID'],axis=1)
print(rbscaler.fit(test_for_scaling))
test_robustscaled = rbscaler.transform(test_for_scaling) #robustscaling 적용
test_robustscaled = pd.DataFrame(test_robustscaled,columns=test_for_scaling.columns)
test_robustscaled = pd.concat([old_test[['id','fiberID']], test_robustscaled], axis=1) 
test_robustscaled.head()
```

```
X_robust=train_robustscaled.drop(columns=['type'],axis=1)
y_robust=train_robustscaled['type']
X_train_robust, X_test_robust, y_train_robust, y_test_robust= train_test_split(X_robust,y_robust,test_size=0.2,random_state=21) 
```

```
cb.fit(X_train_robust,y_train_robust)
rf.fit(X_train_robust,y_train_robust)
xgb.fit(X_train_robust,y_train_robust)
lgbm.fit(X_train_robust,y_train_robust)
print(log_loss(y_true=y_test_robust,y_pred=cb.predict_proba(X_test_robust)))
print(log_loss(y_true=y_test_robust,y_pred=rf.predict_proba(X_test_robust)))
print(log_loss(y_true=y_test_robust,y_pred=xgb.predict_proba(X_test_robust)))
print(log_loss(y_true=y_test_robust,y_pred=lgbm.predict_proba(X_test_robust)))
```
