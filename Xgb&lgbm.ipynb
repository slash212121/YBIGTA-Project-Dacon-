```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./data/train.csv')
test_data = pd.read_csv('./data/test.csv')
data.head()
```

```python
test_data.head(5)
```

```python
len(data['type'].unique())
```

```python
X = data[data.columns[2:]]
y= data['type']
```

```python
import xgboost as xgb
xgb_model = xgb.XGBClassifier(n_estimators=2000,
                        n_jobs=4,
                        max_depth=15,
                        learning_rate=0.05,
                        gamma = 0.02,
                        subsample = 0.9,
                        colsample_bytree=0.9,
                        missing=-999,
                        tree_method='gpu_hist')
```

```python
# xgb_model.fit(X, y) 
```

```python
from lightgbm import LGBMClassifier
lgbm_model = LGBMClassifier(boosting_type='gbdt', objective='binary', num_leaves=10,
                                learning_rate=0.1, n_estimators=2000, max_depth=15,
                                bagging_fraction=0.9, feature_fraction=0.9, reg_lambda=0.2)
```

```python
lgbm_model.fit(X,y)
```

```python
id_for_index = test_data['id']
test = test_data[test_data.columns[1:]]
```

```python
predictions = lgbm_model.predict_proba(X = test)
print(predictions)
```

```python
result = pd.DataFrame(data=predictions, index=id_for_index, columns=lgbm_model.classes_)
result.head()
```

```python
sample = pd.read_csv("./data/sample_submission.csv")
col_order = sample.columns[1:]
```

```python
result = result[col_order]
result.head()
```

```python
result.to_csv('./result/lgbm_clf.csv', sep=',')
```
